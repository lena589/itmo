```python
class Chocolate:
  def __init__(self, cost:float, addings:str=None):
    if cost<0:
      raise ValueError ('цена меньше нуля')
    self.cost = cost
    self.addings = addings

  def show_my_chocolate (self):
      if self.addings:
        print (f'Шоколадный батончик, добавка: {self.addings}')
      else:
        print ('Обычный шоколадный батончик')

  def __str__(self):
    return f"Шоколадный батончик, {self.cost} рублей"
```

```python
bar = Chocolate (4)
bar.show_my_chocolate()
print (bar)
```

```python
bar = Chocolate (-4)
print (bar)
bar.show_my_chocolate()
```

```python
bar = Chocolate (4, 'nuts')
bar.show_my_chocolate()
print (bar)
```
